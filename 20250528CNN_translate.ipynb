```python
import requests
from lxml import etree
from newspaper import Article
from deep_translator import GoogleTranslator
import pandas as pd
from datetime import datetime
import time

headers = {
    "User-Agent": ("Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                   "AppleWebKit/537.36 (KHTML, like Gecko) "
                   "Chrome/122.0.0.0 Safari/537.36")
}
url = "https://edition.cnn.com/world"

#start_date = datetime(2024, 2, 1)
#end_date = datetime(2024, 2, 29)

res = requests.get(url, headers=headers, timeout=10)
res.encoding = "utf-8"
html = etree.HTML(res.text)
#labels = html.xpath('//div[@class="container_lead-plus-headlines__cards-wrapper"]')  

labels = html.xpath('//div[@class="container__field-links container_lead-plus-headlines__field-links"]') 
titles, links, dates = [], [], []

for div in labels:
    # 获取标题
    title = div.xpath('.//div[@class="container__headline container_lead-plus-headlines__headline"]/span/text()')  
    #title = div.xpath('.//div/a/[class="container__link container__link--type-article container_lead-plus-headlines__link"]/@data-zjs-card_name')
    #title = div.xpath('.//a[contains(@class, "container__link container__link--type-article container_lead-plus-headlines__link")]/text()')
    if title:
        titles.append(title[0].strip())

    # 获取链接
    #link = div.xpath('.//a/@href')  
    link = div.xpath('.//div/@data-open-link') 
    
    if link:
        links.append("https://edition.cnn.com" + link[0])  # 拼接完整链接
    else:
        links.append("")

    # 如果需要日期，可以取消下面的注释并调整正确的日期xpath
    # date = div.xpath('.//span[@class="date"]/text()')
    # if date:
    #     dates.append(date[0].strip())

print(links)
print(titles)
```

```python
assert len(titles) == len(links) 
#==len(dates)

df = pd.DataFrame({
    #"Date": dates,
    'Website': links,
    "NewsTitle": titles
})

df.to_csv("news1.csv", index=False, encoding="utf_8_sig")
print("✓ 已保存到 news1.csv，共抓取 {} 条新闻".format(len(df)))
```

```python
news_data = []

for url in links:
    try:
        article = Article(url)
        article.download()
        article.parse()
        
        title_en = article.title
        text_en = article.text
        
        title_zh = GoogleTranslator(source='auto', target='chinese (simplified)').translate(title_en)
        text_zh = GoogleTranslator(source='auto', target='chinese (simplified)').translate(text_en)
        
        news_data.append({
            'yingwenbiaoti': title_en,
            'zhongwenbiaoti': title_zh,
            'yingwenzhengwen': text_en,
            'zhongwenzhengwen': text_zh,
            'URL': url
        })
        
        print(f"Successfully done: {title_en}")
        
        time.sleep(10)
    
    except Exception as e:
        print(f"Error with URL: {url}\n{e}")

df = pd.DataFrame(news_data)
df.to_csv('news_articles.csv', index=False, encoding="utf-8-sig")

print("Data has been successfully saved to 'news_articles.csv'.")
```
